## Obtain the spin and symmetry

```python
import pandas as pd
f_path = f_qdetout[0]
def Read_qdet(f_path):
    """
    This method will return two dataframe:
    Singlet solution and triplet solution
        return: df_triplet, df_singlet
    
    """
    df = pd.read_csv(f_path)
    #print("read in df: \n",df)
    #df['Sym_label']=df['Sym'].str.extract(r'(\w+)\(.+\)')
    #df["Sym_comp"]=df['Sym'].str.extract(r'\w+\((.+)\)')
    #print("extract symmetry : \n",df)
    df_triplet = df[round(df['Spin'],0)==3]
    df_singlet = df[round(df['Spin'],0)==1]
    #print("triplet part: \n",df_triplet)
    #print("singlet part: \n",df_singlet)
    return df_triplet, df_singlet
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
############
plt.rcParams["lines.linewidth"] = 2
plt.rcParams["xtick.labelsize"] = 15
plt.rcParams["ytick.labelsize"] = 15
plt.rcParams["font.size"] = 15
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = "Arial"
plt.rcParams["axes.titlesize"] = 15
#plt.rcParams["legend.fontsize"] = 16
#plt.rcParams["figure.figsize"] = (12,5.5)
plt.rcParams["figure.figsize"] = (13,6)
plt.rcParams["figure.dpi"] = 100
#_____
ylim=(-0.2,5)
def Plot_Energy_unpol(ax,df,ylim=(-0.2,5),i_mark=[],i_label=[],i_color='black'):
    """
    df : dataframe of qdet output
    """
    Spin = round(df['Spin'],0).tolist()
    E = df['Energy'].tolist()
    #print(Spin)
    if len(i_mark):    
        if isinstance(i_color, str):
            i_color = [i_color]*len(i_mark) # it the color is str, then all label have same color
        for index, i in enumerate(i_mark):#index is the index of element, i is the elment number
            y=[E[i], E[i]]
            if Spin[i]==1:
                x = [0.45,0.62]
            else:
                x = [0.18,0.35]
            ax.plot(x, y, color=i_color[index])
            if y[0]<=ylim[1]: #add the text to the marked band
                last_i = i_mark[index-1] #the last marked state
                if abs(E[i]-E[last_i])<0.1:
                    ax.text(x[0]-0.05,y[0]+0.25,i_label[index],fontsize=12,color=i_color[index])
                    ax.text(x[0]+0.08,y[0]+0.25,round(E[i],2),fontsize=12,color=i_color[index])
                else:
                    ax.text(x[0]-0.05,y[0]+0.05,i_label[index],fontsize=12,color=i_color[index])
                    ax.text(x[0]+0.08,y[0]+0.05,round(E[i],2),fontsize=12,color=i_color[index])
    for i in range(0,len(E)):
        if i not in i_mark:
            y=[E[i], E[i]]
            if Spin[i]==1:
                x = [0.45,0.62]
            else:
                x = [0.18,0.35]
            ax.plot(x, y, color='grey',alpha=0.6)
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.set_xlim([0.1, 0.7])
#Now set the marked states:
mark05 = [0,1,3,4,6]
label05 = [r"$3A'_2$",r"$1E'$",r"$1A'_1$",r"$3E'$",r"$3E''$"]
mark03 = [0,1,3,4,6]
label03 = [r"$3A'_2$",r"$1E'$",r"$1A'_1$",r"$3E'$",r"$3E''$"]
mark25 = [0,1,3,10,4]
label25 = [r"$3A'_2$",r"$1E'$",r"$1A'_1$",r"$3E'$",r"$3E''$"]
mark02 = [0,1,3,10,4]
label02=[r"$3A'_2$",r"$1E'$",r"$1A'_1$",r"$3E'$",r"$3E''$"]
mark15=[0,1,3,4,6]
label15=[r"$3A'_2$",r"$1E'$",r"$1A'_1$",r"$3E'$",r"$3E''$"]
Mark = [mark05,mark03,mark25,mark02,mark15]
Label = [label05,label03,label25,label02,label15]
fig, ax = plt.subplots(nrows=1, ncols=5,sharey=True)
for i in range(len(ax)):
    f_path = f_qdetout[i]
    df = pd.read_csv(f_path)
    #df_triplet,df_singlet = Read_qdet(f_path)
    ax[i].set_ylim(ylim)
    Plot_Energy_unpol(ax[i],df,i_mark = Mark[i],i_label = Label[i])
    ax[i].set_title(f"L>{Thro[i]}({Orb[i]}o)")
ax[0].set_ylabel("Energy(eV)")
fig.subplots_adjust(wspace=0)
plt.savefig("Spinunpol_QDET_Converge.pdf")

```

# Energy change in line plot


```python
Label=[r"$3A'_2$",r"$1E'$",r"$1A'_1$",r"$3E'$",r"$3E''$"] # a list of interested states
#Here are the energy index of each states
L = [0.5,0.3,0.25,0.2,0.15,0.13,0.12] # a list of threshold
Orb = [9, 15, 24, 37, 45, 51, 59] #number of orbitals related to the threshold
i_3Ap2 = [0,0,0,0,0,0,0]
i_1Ep = [1, 1, 1, 1, 1, 1, 1]
i_1Ap1 = [3, 3, 3, 3, 3, 3, 3]
i_3Ep = [4, 4, 10, 10, 4, 10, 4]
i_3Epp = [6, 6, 4, 4, 6, 4, 6]
States= [i_3Ap2, i_1Ep, i_1Ap1, i_3Ep, i_3Epp]
### Read in qdet results:
Dir_name=[f"./L_{thro}" for thro in L]
print(Dir_name)
f_qdetout=[f"{d}/QDET_output.csv" for d in Dir_name]
print(f_qdetout)
E = [] #here E[i] will be the qdet solution corresponded to L[i] threshold
for f in f_qdetout:
    e =pd.read_csv(f)['Energy'].tolist() #the energy as a list
    E.append(e)
############
plt.rcParams["lines.linewidth"] = 2
plt.rcParams["xtick.labelsize"] = 15
plt.rcParams["ytick.labelsize"] = 15
plt.rcParams["font.size"] = 15
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = "Arial"
plt.rcParams["axes.titlesize"] = 15
#plt.rcParams["legend.fontsize"] = 16
#plt.rcParams["figure.figsize"] = (12,5.5)
plt.rcParams["figure.figsize"] = (6,6)
plt.rcParams["figure.dpi"] = 100
#_____
ylim=(-0.2,6)
fig, ax = plt.subplots(nrows=1, ncols=1)
def plot_state(ax, L, E, i_state, lab, color='black'):
    """
    plot the certain manybody state vs localizarion factor

    ax: pyplot axis object
    L : list of string. 
        act as the x label, can be Localization threshold or orbital numbers
    E: a list of list. 
        where E[i] is the qdet solution for ith localization factor
    i_state: list of int index
        index of certein state in the qdet solution
    lab: 
        label of the certain state to plot
    color: str
        color of the line
    
    """
    y = [] #y should be the energy of this same state at each L
    if not (len(L)==len(E) and len(E)==len(i_state)):
        raise IndexError("length of L, E, i_state should be the same")
    for n, i in enumerate(i_state):
        print(f'state {lab} at L>{L[n]} is {E[n][i]}  ')
        y.append(E[n][i])
    x = np.arange(len(L))+0.5 #x axis
    ax.plot(x,y,marker='o',color=color,label = lab)
    plt.xticks(x,[l for l in L])
Colors=["C0","C1","C2","C3","C4"]
X_tick=[f"{o}o" for o in Orb]
for j,s in enumerate(States): #plot all the states
    plot_state(ax, X_tick, E, s, Label[j],color = Colors[j])
plt.ylabel("Energy(eV)")
plt.ylim(ylim)
plt.legend(loc='upper right',fontsize=12)
plt.savefig("converge_line.pdf")

    
    
```
